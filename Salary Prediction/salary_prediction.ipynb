```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import nltk

# Download WordNet data if not already downloaded
# nltk.download('wordnet')
# nltk.download('omw-1.4')  # For multilingual WordNet
# nltk.download('averaged_perceptron_tagger')  # For POS tagging
# nltk.download('averaged_perceptron_tagger_eng')

lemmatizer = WordNetLemmatizer()

# Function to get WordNet POS tag from nltk POS tag
def get_wordnet_pos(word):
    """Map NLTK POS tag to WordNet POS tag."""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {
        'J': wordnet.ADJ,
        'N': wordnet.NOUN,
        'V': wordnet.VERB,
        'R': wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)

# Lemmatization function
def lemmatize(text):
    lemmatized_words = []
    for word in text.split():
        pos = get_wordnet_pos(word)  # Get the POS tag for each word
        lemmatized_words.append(lemmatizer.lemmatize(word, pos))
    return " ".join(lemmatized_words)
```

```python
tokenized_text = df1['Job Summary'].apply(lambda x : x.split())
word2vec_model = Word2Vec(sentences=tokenized_text,vector_size=100,window=5,min_count=5,workers=4)
def get_Sentence_embeddings(sentence,model):
    return np.mean([model.wv[word] for word in sentence if word in model.wv],axis = 0)
df1['embedding']= tokenized_text.apply(lambda x: get_Sentence_embeddings(x,word2vec_model))
```
